```python
X_test = X.drop(columns={'native','maritalstatus','race','sex'},axis=1)
X_test
```

```python
X_dum = pd.get_dummies(X_test)
```

```python
X_dum.shape
```

```python
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
X_test_norm = SS.fit_transform(X_dum)
X_test_norm
```

```python
# Build a model
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
model = SVC() # default kernal='rbf' and default panalty C=1 
model.fit(X_test_norm,y)
model.score(X_test_norm,y)  #check training accuracy
```
