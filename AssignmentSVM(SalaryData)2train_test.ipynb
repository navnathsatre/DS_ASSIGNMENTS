```python
X_train = X.drop(columns={'native','maritalstatus','race','sex'},axis=1)
```

```python
X_train
```

```python
X_dum = pd.get_dummies(X_train)
```

```python
X_dum.shape
```

```python
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
X_train_norm = SS.fit_transform(X_dum)
X_train_norm
```

```python
# Build a model
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
model = SVC(C=18,kernel='poly',degree=4) # default kernal='rbf' and default panalty C=1 
model.fit(X_train_norm,y_train.values.ravel())
model.score(X_train_norm,y_train.values.ravel())  #check training accuracy
```
