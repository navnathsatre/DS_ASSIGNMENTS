Use Random Forest to prepare a model on fraud data 
treating those who have taxable_income <= 30000 as "Risky" and others are "Good"

```python
#load the data set
import pandas as pd
Fraud_data=pd.read_csv("D:\\DATA SCIENCE\\DATA SCIENCE ASSIGNMENTS\\Assignments\\Random Forest\\Fraud_check.csv")
Fraud_data
```

```python
data=Fraud_data.copy()
```

```python
data.rename(columns={'Undergrad':'Ugrad','Marital.Status':'Mstatus','Taxable.Income':'Taxinc','City.Population':'Citypopu','Work.Experience':'Wexp'},inplace=True)
```

```python
df=data.copy()
```

```python
df['Ugrad'].replace({'NO':0,'YES':1},inplace=True)
df['Mstatus'].replace({'Single':0,'Married':1,'Divorced':2},inplace=True)
df['Urban'].replace({'NO':0,'YES':1},inplace=True)
```

```python
df
```

```python
# Check outliers
from sklearn.ensemble import IsolationForest
```

```python
clf=IsolationForest(contamination=0.01,random_state=99)
clf.fit(df)
```

```python
df['score']=clf.decision_function(df)
df['anomaly']=clf.predict(df.iloc[:,0:6])
```

```python
df
```

```python
df[df['anomaly']==-1]
```

```python
df=df.drop(df.index[[21,112,168,414,541,574]],axis=0).reset_index()
```

```python
df=df.drop(['index'],axis=1)
```

```python
df
```

```python
Y=df['Taxinc']
Y=[0 if b<=30000 else 1 for b in Y]
Y=pd.DataFrame(Y)
Y.value_counts()   #inbalance Y
```

```python
df1=df.copy()
```

```python
X=df1.drop(columns=['Taxinc','score','anomaly'],axis=1)
X
```

```python
Y.values
```

## Random Forest as Classifier

```python
#model validation technique(train test split)
from sklearn.model_selection import train_test_split
```

```python
X_TRAIN,X_TEST,Y_TRAIN,Y_TEST=train_test_split(X,Y,test_size=0.3,random_state=55)
```

```python
#Build a model
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier(n_estimators=113,max_features=2,random_state=852)
model.fit(X_TRAIN,Y_TRAIN)
model.score(X_TEST,Y_TEST)
```

```python
Y_PRED=model.predict(X_TEST)
from sklearn.metrics import classification_report
print(classification_report(Y_TEST,Y_PRED))
```

```python
#model validation technique(K-Fold)
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
#model validation technique(k-fold)
k_fold=KFold(n_splits=5)
RESULTS=cross_val_score(model,X,Y,cv=k_fold)
RESULTS.mean()
```
